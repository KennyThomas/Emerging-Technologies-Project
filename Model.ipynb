# Model for predicting power from wind speed

### Imports

```python
import matplotlib.pyplot as plt
from sklearn import linear_model, metrics, model_selection
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
from sklearn.model_selection import train_test_split 
import tensorflow.keras as kr
from tensorflow.keras.layers import Dense, Activation,Conv2D,MaxPooling2D,Flatten,Dropout
import pickle
filename='model.pkl'

```

### Read in the data


```python
dataset = pd.read_csv("data.csv")
speed=dataset['speed']  #Get the column speed
power=dataset['power']  #Get the column power

x = np.array(speed).reshape(-1, 1)
y = np.array(power)

#Splitting the data into Train and Test

xtrain, xtest, ytrain, ytest = train_test_split(x,y,test_size=1/5, random_state=1)
```

### Display the data


```python
dataset.plot(x='speed', y='power', style='b.')
plt.title('Speed & power')
plt.xlabel('speed')
plt.ylabel('power')
plt.show()
```

# Models 

# Keras Model

```python
train = pd.DataFrame()
```

### Train the model

```python
model = kr.models.Sequential()
model.add(Dense(1, activation='sigmoid'))
model.add(kr.layers.Dense(1, input_shape=(1,), activation='linear', kernel_initializer="ones", bias_initializer="zeros"))
model.add(Dense(4, input_dim=1, activation='relu'))
model.add(Dense(1, activation='relu'))
opt = kr.optimizers.Adam(lr=0.01)
model.compile(loss='mean_squared_error', optimizer=opt)
model.fit(x, y, epochs=500)
accuracy = model.evaluate(x, y)
print('Accuracy: %.2f' % (accuracy))
```

```python
predictions = model.predict(xtest)
plt.plot(x, y,'b.')
plt.title('Actual')
plt.xlabel('speed')
plt.ylabel('power')
plt.show()

plt.plot(xtest, predictions,'b.')
plt.title('Predicted ')
plt.xlabel('speed')
plt.ylabel('power')
plt.show()

```

 # DecisionTreeRegressor

```python
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor(random_state=1)
regressor.fit(xtrain, ytrain)

            
            
```

### Training data used

```python
plt.plot(xtrain, ytrain,'r.')
plt.title('Training Data')
plt.xlabel('speed')
plt.ylabel('power')
plt.show()
```

### Testing the accuracy


```python

y_pred = regressor.predict(xtest)
df = pd.DataFrame({'Actual': ytest, 'Predicted': y_pred})
df.describe()

```

As you can see here the model is not 100% accurate, but the prediction is very close

```python

plt.plot(x, y,'r.')
plt.title('Actual')
plt.xlabel('speed')
plt.ylabel('power')
plt.show()

plt.plot(xtest, y_pred,'r.')
plt.title('Predicted ')
plt.xlabel('speed')
plt.ylabel('power')
plt.show()
```
