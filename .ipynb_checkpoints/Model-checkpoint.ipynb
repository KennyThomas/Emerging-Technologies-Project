# Model for predicting power from wind speed

### Imports

```python
import matplotlib.pyplot as plt
from sklearn import linear_model, metrics, model_selection
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
from sklearn.model_selection import train_test_split 
import pickle
filename='model.pkl'

```

### Read in the data


```python
dataset = pd.read_csv("data.csv")
speed=dataset['speed']  #Get the column speed
power=dataset['power']  #Get the column power
```

### Display the data


```python
dataset.plot(x='speed', y='power', style='b.')
plt.title('Speed & power')
plt.xlabel('speed')
plt.ylabel('power')
plt.show()
```

### Model is trained using the training data

```python
x = np.array(speed).reshape(-1, 1)
y = np.array(power)

#Splitting the data into Train and Test

xtrain, xtest, ytrain, ytest = train_test_split(x,y,test_size=1/5, random_state=1)


# I chose to use the DecisionTreeRegressor as it was more accurate
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor(random_state=1)
regressor.fit(xtrain, ytrain)

            
            
```

### Trainging data used

```python
plt.plot(xtrain, ytrain,'r.')
plt.title('Speed & power')
plt.xlabel('speed')
plt.ylabel('power')
plt.show()
```

### Testing the accuracy


```python

y_pred = regressor.predict(xtest)
df = pd.DataFrame({'Actual': ytest, 'Predicted': y_pred})
df.describe()
```

As you can see here the model is not 100% accurate, but the prediction is very close

### User input

```python
input_area = float(input("Enter speed: "))
powerEstimate = regressor.predict([[input_area]])
print("power is :", round(powerEstimate[0], 5))
```

The user can enter a wind speed and the model can predict using the training data the power
